```python
forties_ctrl  = np.sum(ctrl_psd[l40:u40])
forties_variants = np.zeros((86,)) 


for j in range(0,76):
    forties_variants[j]  = np.sum(variant_psds[j,l40:u40])

for j in range(0,4):
    forties_variants[j+76]  = np.sum(variant_hcn_psds[j,l40:u40])
    
for j in range(0,6):
    forties_variants[j+80]  = np.sum(variant_scn_psds[j,l40:u40])

print(forties_variants)
gamma_change = forties_variants/forties_ctrl
print(gamma_change)
```

### Calculate correlations

```python
from scipy.stats import pearsonr, spearmanr
```

```python
# convert to float
ds2 = [float(x) for x in ds[indices]]
ppis2 = [float(x) for x in ppis[indices]]
```

```python
gamma_delta_pearson_r,gamma_delta_pearson_p = pearsonr(gamma_change,ds2)
gamma_delta_spearman_r,gamma_delta_spearman_p = spearmanr(gamma_change,ds2)

print('Pearson r: {0} p: {1}'.format(gamma_delta_pearson_r,gamma_delta_pearson_p))
print('Spearman r: {0} p: {1}'.format(gamma_delta_spearman_r,gamma_delta_spearman_p))
```

```python
gamma_ppi_pearson_r,gamma_ppi_pearson_p = pearsonr(gamma_change,ppis2)
gamma_ppi_spearman_r,gamma_ppi_spearman_p = spearmanr(gamma_change,ppis2)

print('Pearson r: {0} p: {1}'.format(gamma_ppi_pearson_r,gamma_ppi_pearson_p))
print('Spearman r: {0} p: {1}'.format(gamma_ppi_spearman_r,gamma_ppi_spearman_p))
```
